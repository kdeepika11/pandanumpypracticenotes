```python
mean= employees["Salary"].mean()
sd= employees["Salary"].std()
mean
sd
```

```python
mean
sd
```

```python
employees["zsalary"]=(employees["Salary"] - mean)/sd
```

```python
employees[["Name", "Salary", "zsalary"]]

def check(a):
    if(a == 0):
        return "Normal Salary"
    if(a < 0):
        return "Below average"
    else:
        return "Above average"
employees["meaning"] = check(employees["zsalary"])
employees[["Name", "Salary", "zsalary","meaning"]]

```
