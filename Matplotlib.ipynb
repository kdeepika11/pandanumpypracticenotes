```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(1,5,11)
y=x**2
x
```

```python
y
```

```python
%matplotlib inline
```

```python
plt.plot(x, y, 'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Learning')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Learning')

plt.subplot(1,2,2)
plt.plot(x, y, 'b*-')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Learning')
```

```python
fig= plt.figure()
axes= fig.add_axes([0.1,0.1,1,1])

axes.plot(x, y, 'r--')
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('Learning')
```

```python
# create 3 subplots and plot on each Axes
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
#for ax in axes:
    #ax.plot(x, y, 'g')
    #ax.set_xlabel('x')
    #ax.set_ylabel('y')
    #ax.set_title('title')
#fig
axes[0].plot(x, y, 'r--')
axes[0].set_xlabel('x axis')
axes[0].set_ylabel('y axis')
axes[0].set_title('Learning1')

axes[1].plot(x, y, 'b--')
axes[1].set_xlabel('x axis')
axes[1].set_ylabel('y axis')
axes[1].set_title('Learning2')

axes[2].plot(x, y, 'g--')
axes[2].set_xlabel('x axis')
axes[2].set_ylabel('y axis')
axes[2].set_title('Learning3')

plt.tight_layout()
```

```python
fig.savefig('chart.png')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for ax in axes:
    ax.plot(x, y, 'g',label='xy')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-', label='x') # blue line with dots
ax.plot(x, x**3, 'g--', label='x')
ax.legend()
```
